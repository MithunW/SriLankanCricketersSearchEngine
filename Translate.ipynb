```python
import json
import os
import re
from mtranslate import translate

json_file = open("/cricketerdata.json", "r", encoding="utf-8")
data = json.load(json_file)

people = []

count = 0
for person in data:
  attrs = person.keys()
  si_person= {}

  for att in attrs:
      txt = person[att].lower()
      lines = txt.split('\n')
      lst = ['']
      res = ''
      for i in lines:
        if len(lst[-1]) + len(i) < 5000:
          lst[-1] += '\n\n' + i
        else:
          lst.append('\n\n' + i)
      for i in lst:
        res += '\n\n' + translate(i, 'si', 'en')
      res = re.sub(r'\n\s*\n', '\n\n', res)
      si_person[att] = res[2:]
  people.append(si_person)
  count += 1
  print("{:0.2f}% Completed".format((count/225)*100))
  

with open('/pm-si-data.json', 'w', encoding='utf8') as f:
    f.write(json.dumps(people, ensure_ascii=False))
```

```python
si_json_file = open("/pm-si-data.json", "r", encoding="utf-8")
si_data = json.load(si_json_file)
```

```python
from collections import defaultdict
```

```python
stat = defaultdict(int)
for i, singer in enumerate(si_data):
  for k, v in singer.items():
    stat[k] += 1
sorted_stat = sorted(stat.items(), key=lambda x: -x[1])
for k, v in sorted_stat:
  print(k.ljust(20), str(v).rjust(3))

fields = [k for k, v in sorted_stat[:10]]
print(fields)
```

```python
filtered_cricketers = []
drop_fields = [ "From", "Type", "Intro", "A.K.A."]
selected_fields = ["cricketerName","bio","paragraph", "Gender", "Is", "Birth", "Age", "Star sign", "Star sign","Sports Teams","Death","Profiles"]
for cricketer in si_data:
  cricketer_temp= cricketer.copy()
  for k in cricketer.keys():
    if k in drop_fields:
      cricketer_temp.pop(k)
  

  for field in selected_fields:
    if field in cricketer_temp.keys():
      print("oK")

    else:
      cricketer_temp[field]=""
  filtered_cricketers.append(cricketer_temp)

```

```python
with open('/filtered-si-data.json', 'w', encoding='utf8') as f:
    f.write(json.dumps(filtered_cricketers, ensure_ascii=False))
```
